# Web scraping - Beautiful Soup
Web scraping is having a computer work with web pages through their html, css, etc.

This is the only option for working with websites when there is no API or of the API does not give the needed options.

```
from IPython.display import Image, display

Image('https://upload.wikimedia.org/wikipedia/commons/thumb/d/d0/German_Shepherd_-_DSC_0346_%2810096362833%29.jpg/440px-German_Shepherd_-_DSC_0346_%2810096362833%29.jpg', width=100)
```

```
import requests
from bs4 import BeautifulSoup

response = requests.get('https://en.wikipedia.org/wiki/German_Shepherd')

soup = BeautifulSoup(response.text, 'html.parser')

for img in soup.find_all('img')[:4]:
  full_img = 'https:' + img['src']
  if full_img.endswith('jpg'):
    display(Image(full_img, width=100))
```

#Relative and Absolute paths
As we parse web pages, some images will start with / which gets appended to the root url, while others start with // which define an entire url by itself. 

We started with conditionals to check for both, but we ended up switching to urllib.

```
from urllib.parse import urljoin
url = 'https://en.wikipedia.org/wiki/PlayStation_5'
path = '/static/images/footer/wikimedia-button.png'
print(urljoin(url, path))

url = 'https://en.wikipedia.org/wiki/PlayStation_5'
path = '//upload.wikimedia.org/wikipedia/commons/thumb/a/a1/Mark_Cerny%2C_Gamelab_2018_%2843111149312%29.jpg/170px-Mark_Cerny%2C_Gamelab_2018_%2843111149312%29.jpg'
print(urljoin(url, path))
```

We showed an alternative way of working with file types using mimetypes.

```
import mimetypes
import re
from urllib.parse import urljoin

def print_images(url, start=None, stop=None):
  response = requests.get(url)

  soup = BeautifulSoup(response.text, 'html.parser')

  for img in soup.find_all('img')[start:stop]:
    full_img = urljoin(url, img['src'])
    mimetype = mimetypes.guess_type(full_img)
    if mimetype[0] is None:
      continue
    if mimetype[0] in ['image/jpeg', 'image/png']: #gif wouldn't work with Image
      display(Image(full_img, width=100))

url = input('Give me a web address: ')
#start = int(input('Give me a starting index: '))
#stop = int(input('Give me a stopping index (exclusive)'))
print_images(url)
```

#The final code

```
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

def print_images(url):
  response = requests.get(url)

  soup = BeautifulSoup(response.text, 'html.parser')

  for img in soup.find_all('img'):
    full_img = urljoin(url, img['src'])
    if full_img.endswith(('jpg', 'png')):
      display(Image(full_img, width=100))

url = input('Give me a web address: ')
print_images(url)
```

# Dates and times
Next up we are going to get some practice working with datetime. This can be used to keep track of specific dates as well as lengths of time (such as how far two dates are apart).

```
from datetime import datetime

now = datetime.now()
print(now)

future = datetime(2021, 1, 1)
print(future)

time_dif = future-now
print(f'Time until 2021: {time_dif}')

hours = time_dif.seconds // 3600
minutes = time_dif.seconds % 3600 // 60
seconds = time_dif.seconds - (hours * 3600) - (minutes * 60)
print(time_dif.days, hours, minutes, seconds, time_dif.microseconds)
```

```
now = datetime.now()
year = now.year

future = datetime(year+1, 1, 1)
print(future-now)
```

## Shallow copy

```
from copy import copy
data1 = [5]
#data2 = data1[:]
#data2 = data1.copy()
data2 = copy(data1)

print(id(data1), id(data2))

data1[0] = 6

print(data1, data2)
print(id(data1), id(data2))
```

```
def print_list(data):
  for d in data:
    print(id(d))
  print()

data1 = ['0', 1, 2, 3, [4], {5}, {6:7}]
data2 = copy(data1)

print_list(data1)
print_list(data2)
```

```
from copy import deepcopy

def print_list(data):
  for d in data:
    print(id(d))
  print()

data1 = ['0', 1, 2, 3, [4], {5}, {6:7}]
data2 = deepcopy(data1)

print_list(data1)
print_list(data2)

data1[4][0] = 'new'
print(data1, data2)

print_list(data1)
print_list(data2)
```

# pprint



```
from pprint import pprint

grades = {
    1 : {
        'math': [90, 90, 30, 20],
        'science': [94, 93, 92, 91],
        'history': [12, 23, 34, 54],
        'reading': [54, 23, 23, 54]
    }, 
    2: {
        'math': [9, 9, 3, 2],
        'science': [9, 9, 9, 9],
        'history': [1, 2, 3, 5],
        'reading': [5, 2, 2, 5]  
    }, 
    3: {
        'math': [45, 30, 20, 0],
        'science': [4, 3, 2, 41],
        'history': [42, 33, 34, 54],
        'reading': [54, 23, 33, 54]
    }
}

pprint(grades)
```

```
import json

print(json.dumps(grades, indent=4))
```

```
with open('data.json', 'w', encoding='utf-8') as f:
  json.dump(grades, f, ensure_ascii=False, indent=4)
```

```
with open('data.json') as f:
  grades = json.load(f)

print(type(grades))
pprint(grades)
```

# Arrays and Collections

```
from array import array

data = array('i', [1, 5, 23, 65, 23, 2])
print(data)

for d in data:
  print(type(d))
```

```
import numpy as np

data = np.array([4, 5, 6, 34, 23, 54, 34, 2])

print(data)
print(np.sum(data))
print(np.sqrt(data))
```

```
data = [3, 4, 5]
data.insert(0, 6)
print(data)
print(data.pop())
print(data.pop())
print(data.pop())
print(data.pop())
```

```
from collections import deque

data = deque([1, 2, 3])

data.appendleft(-7)
print(data.popleft())

print(data)
```
